```python
# ===============================================
# SENTINEL AI - Firearm Detection with Experiments
# AlexNet / ResNet + Confusion Matrices in W&B
# ===============================================

!pip install -q wandb

import os
import time
import copy
import warnings
import random
import numpy as np

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from google.colab import drive
import wandb

warnings.filterwarnings("ignore")

# -------------------------------
# 1Ô∏è‚É£ Mount Drive & Base Settings
# -------------------------------
drive.mount('/content/drive')

# üîÅ Update this path if your dataset is elsewhere
data_dir = "/content/drive/MyDrive/Sentinel AI/guns"

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("‚úÖ Using device:", device)

# -------------------------------
# 2Ô∏è‚É£ W&B Setup
# -------------------------------
wandb.login()
PROJECT_NAME = "sentinel_ai_firearm_detection"

# -------------------------------
# 3Ô∏è‚É£ Data: transforms + loaders
# -------------------------------
def build_dataloaders(
    data_dir,
    batch_size=32,
    use_augmentation=True,
    input_size=224,
    use_grayscale=False
):
    print("\nüì¶ Building dataloaders...")
    print(f"   - batch_size       = {batch_size}")
    print(f"   - use_augmentation = {use_augmentation}")
    print(f"   - use_grayscale    = {use_grayscale}")
    print(f"   - input_size       = {input_size}")

    extra_tfms = []
    if use_grayscale:
        # Convert to 3-channel grayscale so AlexNet/ResNet still work with 3-channel input
        extra_tfms.append(transforms.Grayscale(num_output_channels=3))

    if use_augmentation:
        train_tfms = transforms.Compose(
            [
                transforms.Resize((input_size, input_size)),
                transforms.RandomHorizontalFlip(),
                # Add more augmentations here if you want (Rotation, ColorJitter, etc.)
                *extra_tfms,
                transforms.ToTensor(),
                transforms.Normalize(
                    [0.485, 0.456, 0.406],
                    [0.229, 0.224, 0.225]
                ),
            ]
        )
    else:
        train_tfms = transforms.Compose(
            [
                transforms.Resize((input_size, input_size)),
                *extra_tfms,
                transforms.ToTensor(),
                transforms.Normalize(
                    [0.485, 0.456, 0.406],
                    [0.229, 0.224, 0.225]
                ),
            ]
        )

    valid_tfms = transforms.Compose(
        [
            transforms.Resize((input_size, input_size)),
            *extra_tfms,
            transforms.ToTensor(),
            transforms.Normalize(
                [0.485, 0.456, 0.406],
                [0.229, 0.224, 0.225]
            ),
        ]
    )

    image_datasets = {
        "train": datasets.ImageFolder(os.path.join(data_dir, "train"), train_tfms),
        "valid": datasets.ImageFolder(os.path.join(data_dir, "valid"), valid_tfms),
    }

    dataloaders = {
        x: DataLoader(
            image_datasets[x],
            batch_size=batch_size,
            shuffle=True if x == "train" else False,
            num_workers=2,
        )
        for x in ["train", "valid"]
    }

    dataset_sizes = {x: len(image_datasets[x]) for x in ["train", "valid"]}
    class_names = image_datasets["train"].classes

    print(f"üìÇ Train size: {dataset_sizes['train']} | Valid size: {dataset_sizes['valid']}")
    print(f"üß† Classes: {class_names}")
    print(f"   Train batches: {len(dataloaders['train'])}, Valid batches: {len(dataloaders['valid'])}")

    return dataloaders, dataset_sizes, class_names

# -------------------------------
# 4Ô∏è‚É£ Model factory: AlexNet + ResNet18
# -------------------------------
def build_model(num_classes, architecture="alexnet", weights="imagenet", freeze_features=True):
    """
    architecture: 'alexnet' or 'resnet18'
    weights: 'imagenet' or None
    freeze_features: whether to freeze the convolutional trunk
    """
    print(f"\nüß† Building model:")
    print(f"   - architecture    = {architecture}")
    print(f"   - weights         = {weights}")
    print(f"   - freeze_features = {freeze_features}")

    if architecture == "alexnet":
        if weights == "imagenet":
            print("   -> Loading AlexNet with ImageNet weights...")
            model = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1)
        else:
            print("   -> Loading AlexNet with RANDOM init...")
            model = models.alexnet(weights=None)

        # Freeze features if requested
        for param in model.features.parameters():
            param.requires_grad = not freeze_features

        # Replace final classifier layer
        num_features = model.classifier[6].in_features
        model.classifier[6] = nn.Linear(num_features, num_classes)

    elif architecture == "resnet18":
        if weights == "imagenet":
            print("   -> Loading ResNet18 with ImageNet weights...")
            model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        else:
            print("   -> Loading ResNet18 with RANDOM init...")
            model = models.resnet18(weights=None)

        # Freeze everything except final fc if requested
        if freeze_features:
            for name, param in model.named_parameters():
                if "fc" in name:
                    param.requires_grad = True
                else:
                    param.requires_grad = False
        else:
            for param in model.parameters():
                param.requires_grad = True

        num_features = model.fc.in_features
        model.fc = nn.Linear(num_features, num_classes)

    else:
        raise ValueError(f"Unknown architecture: {architecture}")

    model = model.to(device)
    print("‚úÖ Model ready on device:", device)
    return model

# -------------------------------
# 5Ô∏è‚É£ Optimizer factory
# -------------------------------
def build_optimizer(optimizer_name, params, lr):
    print(f"\n‚öôÔ∏è Building optimizer:")
    print(f"   - type = {optimizer_name}")
    print(f"   - lr   = {lr}")

    name = optimizer_name.lower()
    if name == "adam":
        return optim.Adam(params, lr=lr)
    elif name == "sgd":
        return optim.SGD(params, lr=lr, momentum=0.9)
    elif name == "rmsprop":
        return optim.RMSprop(params, lr=lr, momentum=0.9)
    else:
        raise ValueError(f"Unknown optimizer: {optimizer_name}")

# -------------------------------
# 6Ô∏è‚É£ Plotting helper (local, not W&B)
# -------------------------------
def plot_history(history, title_suffix=""):
    epochs = range(1, len(history["train_loss"]) + 1)

    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, history["train_loss"], label="Train Loss")
    plt.plot(epochs, history["val_loss"], label="Val Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Loss {title_suffix}")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, history["train_acc"], label="Train Acc")
    plt.plot(epochs, history["val_acc"], label="Val Acc")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(f"Accuracy {title_suffix}")
    plt.legend()

    plt.tight_layout()
    plt.show()

# -------------------------------
# 7Ô∏è‚É£ Training + validation + Confusion Matrix in W&B
# -------------------------------
def train_and_validate(experiment_cfg, data_dir):
    """
    experiment_cfg example:
    {
        "experiment_name": "alexnet_adam_rgb_bs32_lr1e-4",
        "architecture": "alexnet",
        "weights": "imagenet",
        "freeze_features": False,
        "optimizer": "adam",
        "epochs": 5,
        "lr": 1e-4,
        "batch_size": 32,
        "use_augmentation": True,
        "use_grayscale": False
    }
    """

    print("\n==============================")
    print("üîß Starting experiment with config:")
    for k, v in experiment_cfg.items():
        print(f"   {k}: {v}")
    print("==============================")

    # 1) Build dataloaders for THIS experiment
    dataloaders, dataset_sizes, class_names = build_dataloaders(
        data_dir=data_dir,
        batch_size=experiment_cfg["batch_size"],
        use_augmentation=experiment_cfg["use_augmentation"],
        input_size=224,
        use_grayscale=experiment_cfg.get("use_grayscale", False),
    )

    # 2) W&B run
    run = wandb.init(
        project=PROJECT_NAME,
        name=experiment_cfg["experiment_name"],
        config=experiment_cfg,
        reinit=True,
    )
    config = wandb.config

    # 3) Build model
    model = build_model(
        num_classes=len(class_names),
        architecture=config.architecture,
        weights=config.weights,
        freeze_features=config.freeze_features,
    )

    criterion = nn.CrossEntropyLoss()

    # Train only trainable params (some may be frozen)
    params_to_update = [p for p in model.parameters() if p.requires_grad]
    print(f"üß© Number of trainable parameter tensors: {len(params_to_update)}")

    optimizer = build_optimizer(config.optimizer, params_to_update, lr=config.lr)
    scheduler = lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

    best_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())

    train_loss_hist, val_loss_hist = [], []
    train_acc_hist, val_acc_hist = [], []

    since = time.time()
    num_epochs = config.epochs

    for epoch in range(num_epochs):
        current_lr = optimizer.param_groups[0]["lr"]
        print(f"\n[{config.experiment_name}] Epoch {epoch+1}/{num_epochs} (lr={current_lr:.6f})")
        print("-" * 40)

        for phase in ["train", "valid"]:
            if phase == "train":
                model.train()
            else:
                model.eval()

            print(f"‚ñ∂ Phase: {phase}")
            running_loss = 0.0
            running_corrects = 0

            num_batches = len(dataloaders[phase])

            for batch_idx, (inputs, labels) in enumerate(dataloaders[phase]):
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == "train"):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == "train":
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

                # üí¨ occasional batch print
                if batch_idx % max(1, num_batches // 4) == 0:
                    print(
                        f"   - {phase} batch {batch_idx+1}/{num_batches} "
                        f"| batch_loss: {loss.item():.4f}"
                    )

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            if phase == "train":
                train_loss_hist.append(epoch_loss)
                train_acc_hist.append(epoch_acc.item())
            else:
                val_loss_hist.append(epoch_loss)
                val_acc_hist.append(epoch_acc.item())

            print(f"‚úÖ {phase} Loss: {epoch_loss:.4f} | Acc: {epoch_acc:.4f}")

            # üìä Log to W&B
            wandb.log(
                {
                    "epoch": epoch + 1,
                    f"{phase}/loss": epoch_loss,
                    f"{phase}/accuracy": epoch_acc.item(),
                    "learning_rate": current_lr,
                }
            )

            # Track best model on validation
            if phase == "valid" and epoch_acc.item() > best_acc:
                best_acc = epoch_acc.item()
                best_model_wts = copy.deepcopy(model.state_dict())
                save_name = f"sentinel_ai_{config.experiment_name}_best.pth"
                torch.save(model.state_dict(), save_name)
                print(f"üíæ New best model saved! Acc: {best_acc:.4f} ({save_name})")

        scheduler.step()

    time_elapsed = time.time() - since
    print(
        f"\n‚è± Training complete in {time_elapsed//60:.0f}m {time_elapsed%60:.0f}s "
        f"| Best val Acc: {best_acc:.4f}"
    )

    # Load best weights before evaluation / confusion matrix
    model.load_state_dict(best_model_wts)

    # -------------------------------
    # üî¢ Confusion matrix on VALID set
    # -------------------------------
    print("\nüìä Computing confusion matrix on validation set...")

    all_preds = []
    all_labels = []

    model.eval()
    with torch.no_grad():
        for inputs, labels in dataloaders["valid"]:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    all_labels = np.array(all_labels)
    all_preds = np.array(all_preds)

    # Raw confusion matrix for console (rows=true, cols=pred)
    num_classes = len(class_names)
    cm_raw = np.zeros((num_classes, num_classes), dtype=int)
    for t, p in zip(all_labels, all_preds):
        cm_raw[t, p] += 1

    print("Confusion matrix (rows = true class, cols = predicted class):")
    print(class_names)
    print(cm_raw)

    # W&B confusion matrix
    cm_wandb = wandb.plot.confusion_matrix(
        y_true=all_labels,
        preds=all_preds,
        class_names=class_names
    )
    wandb.log({"confusion_matrix": cm_wandb})

    wandb.log({"best_val_accuracy": best_acc})
    wandb.finish()

    history = {
        "train_loss": train_loss_hist,
        "val_loss": val_loss_hist,
        "train_acc": train_acc_hist,
        "val_acc": val_acc_hist,
    }
    return model, history

# -------------------------------
# 8Ô∏è‚É£ Experiment Grid
# -------------------------------
architectures = ["alexnet", "resnet18"]
optimizers = ["adam", "sgd"]          # you can add "rmsprop"
learning_rates = [1e-4, 1e-3]
batch_sizes = [16, 32]
use_grayscales = [False, True]       # False = RGB, True = grayscale
use_augmentations = [False, True]

EXPERIMENTS = []
exp_id = 0

for arch in architectures:
    for opt in optimizers:
        for is_gray in use_grayscales:
            exp_id += 1
            cfg = {
                "experiment_name": f"{arch}_{opt}_{'gray' if is_gray else 'rgb'}_exp{exp_id}",
                "architecture": arch,
                "weights": "imagenet",          # change to None for untrained experiments
                "freeze_features": False,       # set True if you want to freeze conv trunk
                "optimizer": opt,
                "lr": random.choice(learning_rates),
                "batch_size": random.choice(batch_sizes),
                "use_augmentation": random.choice(use_augmentations),
                "use_grayscale": is_gray,
                "epochs": 5,
            }
            EXPERIMENTS.append(cfg)

            # Limit total experiments for now (you can raise this)
            if exp_id >= 6:
                break
        if exp_id >= 6:
            break
    if exp_id >= 6:
        break

print("\nüìã Configured experiments:")
for cfg in EXPERIMENTS:
    print(" -", cfg["experiment_name"], "|",
          cfg["architecture"], "|",
          cfg["optimizer"], "|",
          "gray" if cfg["use_grayscale"] else "rgb",
          "| bs", cfg["batch_size"], "| lr", cfg["lr"])

# -------------------------------
# 9Ô∏è‚É£ Run All Experiments
# -------------------------------
all_histories = {}

for cfg in EXPERIMENTS:
    print(f"\nüöÄ Running Experiment: {cfg['experiment_name']}")
    model, history = train_and_validate(cfg, data_dir)
    all_histories[cfg["experiment_name"]] = history
    plot_history(history, title_suffix=f"({cfg['experiment_name']})")

print("\n‚úÖ All Sentinel AI experiments complete!")
print("üîç Compare runs and confusion matrices in your W&B project dashboard.")
```
